Akhirnya, pesawatnya sudah siap dan Anda berangkat ke planet $\Omega$. Setelah melihat secara langsung, Anda sangat terkejut karena ternyata keadaannya jauh berbeda dengan apa yang sudah Anda simulasikan.

* *Martians Eta* ternyata adalah spesies yang kanibal. Faktor kanibalisme diberikan menurut $$C(\eta) = r\times \eta \times \frac{\eta}{\eta+z},$$ dengan $r$ sebagai laju kanibalisme.

Untuk menyesuaikan dengan situasi di lapangan, teman Anda dari Matematika ITB memberikan model mangsa-pemangsa yang baru, biasa disebut sebagai model Holling-Tanner Tipe II dengan modifikasi sebagai berikut

\begin{align*}
  \frac{d\varepsilon}{dt} &= a\varepsilon \left( 1 - \frac{\varepsilon}{K} \right) - b \left(\frac{\varepsilon \eta}{1+m\varepsilon}\right) - \lambda \varepsilon \\[7pt]
  \frac{d\eta}{dt} &= -d \eta + cb\left(\frac{\eta\varepsilon}{1+m\varepsilon}\right) - C(\eta)\\
\end{align*}

di mana $\eta$ menyatakan populasi *martians Eta* dan $\varepsilon$ menyatakan populasi *martians Epsi*.

Konstanta pada sistem persamaan tersebut diperoleh melalui pengamatan yang disajikan pada tabel berikut.

\end{array}

#2. RAGNAROK : The End of an Era

Dalam mitologi Nordik, terkenal $Ragnarok$ merupakan pertarungan akhir untuk suatu peradaban. Berbagai ras di dunia akan berkompetisi dalam memperebutkan sumber daya untuk anak cucu nanti. Terkenal pada zaman dahulu, Manusia dan Dewa
ialah ras yang sangat mendominasi peradaban ini. Perang tak terelakan lagi, dipercaya populasi awal manusia dan dewa berturut-turut adalah $200.000$  $(N_{1})$ dan $75.000$ $(N_{2})$ .

Terbatas nya sumber daya yang ada didunia memaksa Manusia hanya dapat bertahan hidup dengan kuantitas $25.000$ $(K_{1})$ orang. Sedangkan, Dewa dapat bertumbuh hingga $150.000$ $(K_{2})$.

Keterbatasan tersebut mengakibatkan pertumbuhan populasi kedua ras bertumpu terhadap persamaan ***Interspecific Competition*** sebagai berikut,

$$
\frac{dN_{1}}{dt} = \psi_{1}N_{1} \left( \frac{K_{1}-N_{1}-N_{2}}{K_1}\right)
$$

$$
\frac{dN_{2}}{dt} = \psi_{2}N_{2} \left( \frac{K_{2}-N_{2}-\beta N_{1}}{K_2}\right)
$$

Manusia tidak siap untuk perang ini. Hal tersebut tercermin pada koefisien pertumbuhan manusia hanyalah $0.002$ sementara dewa bertumbuh secara superior dengan koefisien pertumbuhan $0.1$. Selain itu juga, dampak kehadiran populasi dewa terhadap populasi manusia mencapai $30\%$ berbanding jauh dengan kebalikan nya, yaitu $10\%$.

Apakah manusia dapat berkompetisi langsung dengan dewa ? Simulasi Anda yang menentukan !

***Catatan*** : Penentuan variabel $\psi_{1},\psi_{2},\alpha,\beta$ diserahkan kepada pembaca sebagai latihan.

## Butir 1

Buatlah Program untuk mensimulasikan $Ragnarok$ Dalam kurun waktu $6$ Bulan ( $180$ Hari ) saja dengan perubahan populasi terjadi sebanyak $100$ kali per hari !

*Output :* Grafik Populasi **Dua** Ras terhadap waktu dalam kurun waktu yang telah ditentukan. Diharapkan parameter berikut ini tersedia pada grafik anda, yaitu Judul, Label sumbu horizontal, Label sumbu vertikal, Grid On, dan Legenda grafik.

```python
Manusia = 200000
Dewa = 75000
M_Manusia = 25000 #carrying capacity manusia
M_Dewa = 150000 #carrying capacity dewa
birth_manusia = 0.002 #rasio pertumbuhan manusia
birth_dewa = 0.1 #rasio pertumbuhan dewa
dewa_manusia = 0.3 #konstanta pertumbuhan dewa terhadap manusia
manusia_dewa = 0.1 #konstanta pertumbuhan manusia terhadap dewa

t = 0
dt = 100*180

Manusia_array = []
Dewa_array = []
time_array = []

while(t<=dt):
  Manusia_array.append(Manusia)
  Dewa_array.append(Dewa)
  time_array.append(t)
  Manusia_new = Manusia + (birth_manusia*Manusia*((M_Manusia - Manusia - manusia_dewa)/M_Manusia))*1/100
  Dewa_new = Dewa + (birth_dewa*Dewa*((M_Dewa - Dewa - dewa_manusia)/M_Dewa))*1/100
  Manusia = Manusia_new
  Dewa = Dewa_new
  t+=1

plt.plot(time_array,Manusia_array)
plt.plot(time_array,Dewa_array)
plt.legend(["Manusia", "Dewa"])
plt.title("Populasi Manusia dan Dewa terhadap Waktu")
plt.ylabel("Populasi")
plt.xlabel("Waktu")
plt.grid()
plt.show()
```

Peperangan panjang masih berlanjut, sayangnya suatu ras mendominasi kompetisi. Penulis mitologi ini merupakan seorang penganut ras tersebut, hal itu menyebabkan ia menutup pena atas tragedi setelahnya.

Dunia bawah terguncang atas menggema nya kabar peperangan akhir abad, hal ini menyebabkan semua iblis haus akan darah serta tertarik untuk berkompetisi dengan Manusia dan Dewa. Iblis bukanlah sosok yang ceroboh, ia membangun rencana $6$ bulan lama nya agar dapat mendominasi $Ragnarok$.

Pertambahan suatu ras baru dalam  ***Interspecific Competition*** tentu akan mempengaruhi ras lain nya. Andaikan terdapat $n$ ras, maka pertumbuhan populasi ras ke-$k$ mengikuti persamaan sebagai berikut

$$
\frac{dN_{k}}{dt} = \psi_{k}N_{k} \left(1- \frac{\sigma_1 N_{1}}{K_k} - \frac{\sigma_{2} N_{2}}{K_k} - ... -\frac{\sigma_{k-1} N_{k-1}}{K_k} - \frac{N_k}{K_k} -\frac{\sigma_{k+1} N_{k+1}}{K_k} - ... -\frac{\sigma_{n} N_{n}}{K_k} \right)
$$

Dengan $\sigma_{i}$ , $\forall (i) \in \{ 1,2,...,k-1,k+1,...,n\}$ merupakan koefisien dampak populasi ras lain terhadap ras $k$.  

Tentunya dengan bergabung nya iblis dalam peperangan kali ini mengakibatkan populasi Manusia dan Dewa terancam. Koefisien dampak populasi Manusia dan Dewa pun berlaku pada iblis, sementara iblis menyerang dengan koefisien mencapai $\gamma=50\%$ terhadap kedua ras lawan nya.
