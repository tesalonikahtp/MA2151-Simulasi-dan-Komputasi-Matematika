```python
import numpy as np
import matplotlib.pyplot as plt

MEC = 10 #batas minimum efektif
MTC = 300 #batas maksimum sebelum keracunan
half_life = 2
interval = 6 #selang waktu antar dosis dalam jam
dosage = (4000/8)*2*1000 #dosis
volume = 3000 #volume plasma 3000mL
absorption_fraction = 0.7
elimination_constant = np.log(2)/half_life #asumsi laju eliminasi drug_in_system sebanding dengan drug_in_system
drug_in_system = 0 #konsentrasi obat di tubuh BJAY
dt = 0.01
t = int(input("Masukkan t : ")) #start waktu pemberian dosis awal

T = [] #array kosong untuk waktu
C = [] #array kosong untuk konsentrasi
ulang = 0

while(ulang<=(24/dt)):

  if(ulang)%(interval//dt) == 0:
    ingested = absorption_fraction*dosage #ingested berarti aliran obat ke dalam drug_in_system
  else :
    ingested = 0
  eliminated = elimination_constant*drug_in_system
  drug_in_system += (ingested-eliminated)*dt
  consentration = drug_in_system/volume
  C.append(consentration*100) #memasukkan konsentrasi dalam array kosong
  ulang+=1
  T.append(ulang*0.01) #memasukkan konsentrasi dalam array kosong

plt.plot(T,C)
plt.title("Grafik Kandungan Paracetamil dalam Tubuh Jay")
plt.xlabel("Waktu (jam)")
plt.ylabel("Konsentrasi Kandungan Paracetamil (mikrogram/mL)")
plt.axhline(10) #menunjukkan batas efektif obat
plt.axhline(300)
```
